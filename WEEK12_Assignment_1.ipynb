```python
# 모듈 불러오기
import pandas as pd
import numpy as np
from scipy.spatial import distance
from geopy.distance import geodesic
import folium
```

```python
# 데이터 불러오기
df = pd.read_csv('data/Starbucks_Seoul.csv')
```

```python
df.head()
```

```python
df.value_counts('gu_name')
```

```python
# 관악구만 필터링
df_k = df[df['gu_name'] == '관악구']
```

```python
## 유클리디안 패키지
pd.DataFrame(distance.cdist(df_k[['latitude','longitude']],
              df_k[['latitude','longitude']],metric = 'euclidean'))
```

```python
## 유클리디안 패키지
city_gu=distance.cdist(df_k[['latitude','longitude']],
              df_k[['latitude','longitude']],metric = 'cityblock')
```

```python
dist_citi_gu=pd.DataFrame(data= city_gu, columns = df_k['name'])
dist_citi_gu.set_index(df_k['name'],inplace=True)
```

```python
dist_citi_gu
```

```python
df_k
```

```python
#거리를 계산하기 위에 두 좌표를 찍고 계산하는 함수 
def cal_euc_dis(row1, row2):
    pnt1 = np.array([row1['latitude'],row1['longitude']])
    pnt2 = np.array([row2['latitude'],row2['longitude']])
    return np.linalg.norm(pnt1- pnt2)
```

```python
# 기준점을 평균으로 잡고 그 평균과 다른 지점들간의 거리를 시각화
map_center = [df_k['latitude'].mean(), df_k['longitude'].mean()]
#folium Map 함수 지도 시각화
m =folium.Map(location = map_center, zoom_start = 10)

# 모든 위치 다양한 스타벅스 관악구 매장에 마커 추가하고 거리를 계산하기 위한 작업
for idx, row in df_k.iterrows():
    folium.Marker(location = [row['latitude'], row['longitude']], popup=row['name']).add_to(m)
    # 각 데이터셋의 위도, 경도를 위치에 넣어서 반복하고 해당 지점의 이름을 같이 마커로 설정하여 위에 만든 m지도에 추가
    
## 첫 번째 시작할 포인트
start_point = df_k.iloc[0]

# 첫 번째 지점에서 다른 지점ㅈ들의 모든 거리를 계산해서 지도에 표시
for idx,row in df_k.iterrows():
    if start_point['name'] !=row['name']:
        dis =cal_euc_dis(start_point, row)
        folium.PolyLine(locations=[[start_point['latitude'], start_point['longitude']],
                                  [row['latitude'], row['longitude']]],
                                  tooltip=f"{start_point['name']} -> {row['name']}:{dis:.2f} meters").add_to(m)
# 지도를 HTML 파일로 저장
m.save('관악구_스타벅스.html')
m
```
