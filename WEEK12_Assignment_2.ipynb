```python
import pandas as pd
import numpy as np
from scipy.spatial import distance
from geopy.distance import geodesic
import folium
```

필수과제 2

- 거리를 배웠으니 위도, 경도 공공데이터가 존재한다.
- 공공 데이터로 데이터를 분석 및 정리한 후 실제 거리 기반으로 지도 시각화까지 해서 insight 정리하는 형태
- 디테일을 살리기 보다 그동안 배웠던 스킬을 가지고 분석하고 인사이트를 1-2 줄 정리해보세요.

```python
df_entire = pd.read_csv('data/library.csv', encoding='cp949')
df=df_entire[df_entire['구명'] == '강남구']
```

```python
df
```

```python
def calculate_distances(df):
    n = len(df)
    dist_matrix = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            if i != j:
                point1 = (df.loc[i, '위도'], df.loc[i, '경도'])
                point2 = (df.loc[j, '위도'], df.loc[j, '경도'])
                dist_matrix[i, j] = geodesic(point1, point2).km  # 지오데식 거리 계산
                
    return dist_matrix

distance_matrix = calculate_distances(df)
distance_df = pd.DataFrame(distance_matrix, index=df['도서관명'], columns=df['도서관명'])
distance_df
```

```python
import folium

# 지도 생성
map_center = [df['위도'].mean(), df['경도'].mean()]
map = folium.Map(location=map_center, zoom_start=13)

# 도서관 위치 표시
for idx, row in df.iterrows():
    folium.Marker([row['위도'], row['경도']], popup=row['도서관명']).add_to(map)

# 도서관 간 거리 표시
for i in range(len(df)):
    for j in range(i+1, len(df)):
        point1 = (df.loc[i, '위도'], df.loc[i, '경도'])
        point2 = (df.loc[j, '위도'], df.loc[j, '경도'])
        distance = distance_matrix[i, j]
        folium.PolyLine([point1, point2], color='blue', weight=2.5, opacity=1).add_to(map)
        folium.Marker(
            location=[(point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2],
            icon=folium.DivIcon(html=f'<div style="font-size: 10pt">{distance:.2f} km</div>')
        ).add_to(map)

# 지도 저장
map.save("library_map.html")
map
```
