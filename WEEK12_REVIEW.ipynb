```python
# 모듈 불러오기
import numpy as np
from scipy.spatial import distance
```

```python
# 두 점을 만들기
p = np.array([1,2])
q = np.array([5,7])
```

### 1. numpy 이용

```python
# (1) 유클리디안 거리
np.linalg.norm(p-q)
```

```python
# (2) 맨하튼 거리
np.sum(np.abs(p-q))
```

```python
# (3) 민코프스키 거리
def min_dis(p,q,p_value):
    return np.sum(np.abs(p-q)**p_value)*(1/p_value)

min_dis(p, q, 2)
```

```python
# (4) 쳬비쇼프 거리
np.max(np.abs(p-q))
```

```python
# (5) 코사인 유사도
np.dot(p,q)/(np.linalg.norm(p)*np.linalg.norm(q))
```

### 2. scipy 이용

```python
# (1) 유클리디안 거리
distance.euclidean(p, q)
```

```python
# (2) 맨해튼 거리
distance.cityblock(p, q)
```

```python
# (3) 민코프스키 거리
distance.minkowski(p, q)
```

```python
# (4) 쳬비쇼프거리
distance.chebyshev(p, q)
```

```python
# (5) 코사인 유사도
1- distance.cosine(p, q)
```

```python
# 모듈 불러오기
import pandas as pd
from geopy.distance import geodesic
import folium
```

```python
# 데이터 불러오기
df = pd.read_csv('data/Starbucks_Seoul.csv')
```

```python
# 강남구만 필터링
df_g = df[df['gu_name'] == '강남구']
```

```python
## 유클리디안 패키지
pd.DataFrame(distance.cdist(df_g[['latitude','longitude']],
              df_g[['latitude','longitude']],metric = 'euclidean'))
```

```python
## 유클리디안 패키지
city_gu=distance.cdist(df_g[['latitude','longitude']],
              df_g[['latitude','longitude']],metric = 'cityblock')
```

```python
dist_citi_gu=pd.DataFrame(data= city_gu, columns = df_g['name'])
dist_citi_gu.set_index(df_g['name'],inplace=True)
```

```python
#거리를 계산하기 위에 두 좌표를 찍고 계산하는 함수 
def cal_euc_dis(row1, row2):
    pnt1 = np.array([row1['latitude'],row1['longitude']])
    pnt2 = np.array([row2['latitude'],row2['longitude']])
    return np.linalg.norm(pnt1- pnt2)

# 기준점을 평균으로 잡고 그 평균과 다른 지점들간의 거리를 시각화

map_center = [df_g['latitude'].mean(), df_g['longitude'].mean()]
#folium Map 함수 지도 시각화
m =folium.Map(location = map_center, zoom_start = 10)

# 모든 위치 다양한 스타벅스 강남구 매장에 마커 추가하고 거리를 계산하기 위한 작업
for idx, row in df_g.iterrows():
    folium.Marker(location = [row['latitude'], row['longitude']], popup=row['name']).add_to(m)
    # 각 데이터셋의 위도, 경도를 위치에 넣어서 반복하고 해당 지점의 이름을 같이 마커로 설정하여 위에 만든 m지도에 추가
    
## 첫 번째 시작할 포인트
start_point = df_g.iloc[0]

# 첫 번째 지점에서 다른 지점ㅈ들의 모든 거리를 계산해서 지도에 표시
for idx,row in df_g.iterrows():
    if start_point['name'] !=row['name']:
        dis =cal_euc_dis(start_point, row)
        folium.PolyLine(locations=[[start_point['latitude'], start_point['longitude']],
                                  [row['latitude'], row['longitude']]],
                                  tooltip=f"{start_point['name']} -> {row['name']}:{dis:.2f} meters").add_to(m)
# 지도를 HTML 파일로 저장
m.save('강남구_스타벅스.html')
m
```
