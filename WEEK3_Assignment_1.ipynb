- 3가지 칼럼 이상으로 조건을 질문하고 해당 질문에 대한 답변을 query문으로 만들어서 값을 출력
- 최소 10개 이상 질문+코드+답

```python
import seaborn as sns
```

```python
tt=sns.load_dataset('titanic')
tt
```

```python
#Q1:요금이 50이상이고 생존한 여성은 몇명인가?
tt.query('fare>=50 and survived==1 and sex=="female"')
```

A: 82명

```python
#Q2:나이가 20 이상 50이하이고 생존한 여성은 몇명인가?
tt.query('age>=20 and age<=50 and survived==1 and sex=="female"')
```

A:128명

```python
#Q3:요금이 70이상이고 생존한 아이는 몇명인가?
tt.query('fare>=70  and survived==1 and who=="child"')
```

A:5명

```python
#Q4:요금이 20이하이고 생존한 여성의 평균나이는?
tt.query('fare<=20 and survived==1 and sex=="female"').age.mean()
```

A:24.26027397260274

```python
#Q5:요금이 50이하이고 생존하지않은 남성의 평균나이는?
tt.query('fare<=50 and survived==0 and sex=="male"').age.mean()
```

A:30.892857142857142

```python
#Q6:혼자인 여성 중 생존한 인원은 몇명인가?
tt.query('alone==True and sex=="female" and survived==1')
```

A:99명

```python
#Q7:혼자인 남성 중 생존한 인원은 몇명인가?
tt.query('alone==True and sex=="male" and survived==1')
```

A:64명

```python
#Q8: pclass가 1이고 나이가 70이상인 남성은 몇명인가?
tt.query('pclass==1 and age>=70 and sex=="male"')
```

A:4명

```python
#Q9: 생존한 30살이상 남성의 요금 평균은?
tt.query('survived==1 and sex=="male" and age>=30').fare.mean()
```

A:60.32202857142858

```python
#Q10: 생존하지 않은 30살이상 남성의 요금 평균은?
tt.query('survived==0 and sex=="male" and age>=30').fare.mean()
```

A:24.355697701149438
