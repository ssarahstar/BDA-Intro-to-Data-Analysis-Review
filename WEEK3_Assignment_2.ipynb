- 남성과 여성에 따른, 10대 미만, 10대, 20대, 30대 그리고 나머지 그 외에 따라서 fare 평균 값 파생변수 만들기

```python
import seaborn as sns
```

```python
tt=sns.load_dataset('titanic')
```

```python
tt
```

```python
#10대 미만 남성
male_under10_fare_mean=tt.query('age<10 and sex=="male"')['fare'].mean()
male_under10_fare_mean
```

```python
#10대 남성
male_10_fare_mean=tt.query('age>=10 and age<20 and sex=="male"')['fare'].mean()
male_10_fare_mean
```

```python
#20대 남성
male_20_fare_mean=tt.query('age>=20 and age<30 and sex=="male"')['fare'].mean()
male_20_fare_mean
```

```python
#30대 남성
male_30_fare_mean=tt.query('age>=30 and age<40 and sex=="male"')['fare'].mean()
male_30_fare_mean
```

```python
#30대 초과 남성
male_old_fare_mean=tt.query('age>=40 and sex=="male"')['fare'].mean()
male_old_fare_mean
```

```python
#10대 미만 여성
female_under10_fare_mean=tt.query('age<10 and sex=="female"')['fare'].mean()
female_under10_fare_mean
```

```python
#10대 여성
female_10_fare_mean=tt.query('age>=10 and age<20 and sex=="male"')['fare'].mean()
female_10_fare_mean
```

```python
#20대 여성
female_20_fare_mean=tt.query('age>=20 and age<30 and sex=="female"')['fare'].mean()
female_20_fare_mean
```

```python
#30대 여성
female_30_fare_mean=tt.query('age>=30 and age<40 and sex=="female"')['fare'].mean()
female_30_fare_mean
```

```python
#30대 초과 여성
female_old_fare_mean=tt.query('age>=40 and sex=="female"')['fare'].mean()
female_old_fare_mean
```

```python
tt['fare_mean']=0
for i in range(len(tt)):
    if tt.loc[i]['sex']=='female':
        if tt.loc[i]['age']<10:
            tt['fare_mean'].loc[i]=female_under10_fare_mean
        elif tt.loc[i]['age']<20:
            tt['fare_mean'].loc[i]=female_10_fare_mean
        elif tt.loc[i]['age']<30:
            tt['fare_mean'].loc[i]=female_20_fare_mean
        elif tt.loc[i]['age']<40:
            tt['fare_mean'].loc[i]=female_30_fare_mean
        else:
            tt['fare_mean'].loc[i]=female_old_fare_mean
    else:
        if tt.loc[i]['age']<10:
            tt['fare_mean'].loc[i]=male_under10_fare_mean
        elif tt.loc[i]['age']<20:
            tt['fare_mean'].loc[i]=male_10_fare_mean
        elif tt.loc[i]['age']<30:
            tt['fare_mean'].loc[i]=male_20_fare_mean
        elif tt.loc[i]['age']<40:
            tt['fare_mean'].loc[i]=male_30_fare_mean
        else:
            tt['fare_mean'].loc[i]=male_old_fare_mean
        
            
```
