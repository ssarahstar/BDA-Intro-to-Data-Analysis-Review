# query()
- 데이터를 추출하기 위한 조건
- 내가 원하는 데이터를 추출하여서 해당 값을 가져옴

```python
#자료형에러
tt.query('sex==female')
```

```python
#sex(칼럼값)가 "female"인 데이터만 추출
tt.query('sex=="female"')
```

```python
tt.query('age<25')
```

```python
tt.query('alive=="no"')
```

```python
#조건 여러개 사용하기
#생존하고 여성이고 25살 미만
tt.query('''sex=="female" and age<25 and survived==1''')
```

```python
#or 조건도 사용가능
#생존하거나 여성이거나
tt.query('survived==1 or sex=="female"')
```

```python
#tt_fare_mean=tt['fare'].mean()
tt_fare_mean=tt.fare.mean()
tt_fare_mean
```

```python
#변수로는 조건에 넣을 수 없음
tt.query('fare>tt_fare_mean')
```

```python
tt.query('''fare>tt['fare'].mean()''')
```

```python
tt.query('fare>=32 and age<30')[['fare','age']]
```

# sort(정렬)
- 값을 정렬하는 개념
- 오름차순,내림차순 개념
- tt.sort_values('age)

```python
#디폴트값은 오름차순
tt.sort_values('age')
```

```python
#내림차순으로 변경
tt.sort_values('age', ascending=False)
```

```python
tt.sort_values('fare',ascending=False)
```

```python
#두 개 이상의 기준으로 정렬
tt.sort_values(['fare','age'])
#1순위 fare, 2순위 age
```

```python
#기준별로 각각 차순 지정 가능
tt.sort_values(['fare','age'],ascending=[True,False])
```

# 파생변수 만들기
- 기존 변수를 이용해서 새로운 변수를 만드는 것 


```python
#파생변수 초깃값 0으로 세팅 
tt['fare*100']=0
tt
```

```python
#새로운 칼럼의 값으로 fare*100이 출력됨
tt['fare*100']=tt.fare*100
tt
```

```python
#남성의 요금 평균
male_fare_mean=tt.query('sex=="male"')['fare'].mean()
male_fare_mean
```

```python
#여성의 요금 평균
female_fare_mean=tt.query('sex=="female"')['fare'].mean()
female_fare_mean
```

# if문과 for문을 이용하여 전처리 진행
- 남성과 여성의 fare평균값을 새로운 변수로 지정

```python
#새로운 칼럼 생성 후 모두 0으로 초기화
tt['fare_mean']=0
#for 문과 if 문을 이용하여 여성에게는 여성의 요금 평균값을, 남성에게는 남성의 요금 평균값을 부여
for i in range(len(tt)):
    if tt.loc[i]['sex'] == 'female':
        tt['fare_mean'].loc[i]=female_fare_mean
    elif tt.loc[i]['sex']=='male':
        tt['fare_mean'].loc[i]=male_fare_mean
```

```python
#성별에 따른 요금 평균값 지정 확인
tt[['sex','fare_mean']]
```

# 파생변수를 만들 수 있는 문법
- assign(원하는 요구조건)
- lambda

```python
tt
```

```python
#assign(파생변수명 = np.where(요구조건,파생변수값,파생변수 나머지값))
import numpy as np
tt.assign(age_old=np.where(tt['age']>30,'old','young'))
```

```python
tt.iloc[np.where(tt['age']>30)]
```

```python
#np.where은 해당 조건의 인덱스를 반환
np.where(tt['age']>30)
```

```python
#원본데이터프레임은 바뀌지 않음
#assign함수는 기존 데이터프레임을 변경하지 않고 새로운 데이터프레임을 반환함
tt
```

```python
#새로운 변수에 할당
tt_new=tt.assign(age_old=np.where(tt['age']>30,'old','young'))
```

```python
tt_new
```

```python
#lambda
tt.assign(age_100=lambda x:x['age']*100)
```

# groupby()
- 집단들에 대한 그룹을 묶어서 요약통계치등을 보는 문법

```python
#클래스별 나이 평균
tt.groupby('class')['age'].mean()
```

```python
#성별에 따른 요금 평균
tt.groupby('sex')['fare'].mean()
```
