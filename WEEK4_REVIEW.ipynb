```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
#타이타닉 데이터 복사
df=titanic.copy()
```

```python
#pclass칼럼 기준으로 그룹 묶음
pclass_group=df.groupby('pclass')
pclass_group
```

```python
#i: 그룹의 이름을 나타내는 변수, 여기서는 'pclass' 열의 고유값
#j: 각 그룹에 해당하는 데이터를 나타내는 변수, 해당 그룹의 데이터를 포함하는 DataFrame
for i, j in pclass_group:
    print('pclass:',i)
    print(j)
```

```python
#그룹별로 확인하기
pclass_group.get_group(1)
```

```python
pclass_group.get_group(2)
```

```python
pclass_group.get_group(3)
```

```python
#2개 이상 그룹으로 묶기
pcl_sex_group=df.groupby(['pclass','sex'])
```

```python
pcl_sex_group.get_group((1,'female'))
```

```python
#순서가 다르면 에러 
pcl_sex_group.get_group(('male',1))
```

```python
#그룹별로 묶은 상태에서 평균 통계치 
pcl_sex_group.mean()
```

```python
#특정 그룹의 평균 통계치 
pcl_sex_group.get_group((1,'male')).mean()
```

- groupby,agg함수
- agg(요약통계치를 여러 개 확인하거나 내가 원하는 계산식 등을 적용할 때 사용)

```python
pcl_sex_group.agg(['max','min'])
```

```python
pclass_group.agg(['max','min','mean'])
```

```python
#최대와 최소의 차이를 반환하는 함수
def min_max(x):
    return x.max()-x.min()
```

```python
pcl_sex_group.agg(min_max)
```

```python
#평균과 개수의 곱을 반환하는 함수
def meanxcount(x):
    return x.mean()*x.count()
```

```python
pclass_group.agg(meanxcount)
```

```python
#합과 일치하는 지 확인
pclass_group.agg(sum)
```

# 데이터 병합
- 두 개의 데이터 프레임을 붙임
- 두 개를 붙여서 확인해야 하는 상황
- 기준은?데이터 자체에서 고유하게 식별할 수 있는 값

- 판다스에서 제공하는 문법
- merge()
    - 기준 칼럼을 잡고 병합하는 경우
    - 공통의 컬럼을 하나 잡고 병합
- concat()
    - 행기준,열기준으로 병합

# merge
merge(left, right, how='inner', on=None, left_on=None, right_on=None,
             left_index=False, right_index=False, sort=True, suffixes=('_x', '_y'), copy=True, indicator=False)
             
- left: 병합할 때 기준이 될 왼쪽 DataFrame
- right: 병합할 때 함께 결합될 오른쪽 DataFrame
- how: 병합 방식 ('inner', 'outer', 'left', 'right')
- on: 병합할 기준 컬럼의 이름
- left_on과 right_on: 왼쪽과 오른쪽 DataFrame에서 병합할 기준 컬럼의 이름이 다를 때 사용
- left_index와 right_index: 각각 왼쪽과 오른쪽 DataFrame의 인덱스를 병합의 기준으로 사용할지 여부를 지정
- sort: 병합 결과를 기준 컬럼에 따라 정렬할지 여부
- suffixes: 컬럼 이름이 겹칠 때 각각의 DataFrame에 붙일 접미사를 지정
- copy: 병합 결과를 새로운 DataFrame으로 복사할지 여부/ 기본값은 True
- indicator: 병합된 데이터가 어느 DataFrame에서 왔는지를 나타내는 컬럼('_merge')을 추가할지 여부


```python
test1=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['Python','SQL','R','데이터분석입문','데이터분석전처리']})
```

```python
test2=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '상점현황':[4,3,5,10,2]})
```

```python
display(test1)
display(test2)

pd.merge(test1,test2, how='inner')
```

```python
pd.merge(test1,test2, how='left')
```

# concat
concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
- objs: 연결할 데이터
- axis: 연결 방향 지정/ 0은 행 방향(세로), 1은 열 방향(가로). 기본값은 0
- join: 'outer'와 'inner' 중 하나를 선택할 수 있으며, 객체 간의 결합 방식을 결정 'outer'는 합집합,'inner'는 교집합 기본값은 'outer'
- ignore_index: 기본값은 False. True로 설정하면, 연결 후의 인덱스를 기존 인덱스를 무시하고 0부터 시작하는 정수 인덱스로 재설정
- keys: 결과 객체에 대한 다중 인덱스를 생성하는 데 사용될 값/계층적 인덱싱을 구현할 때 유용
- levels: 다중 인덱스 레벨을 지정하는 데 사용
- names: 결과 객체의 인덱스 또는 멀티인덱스의 이름을 지정
- verify_integrity: True로 설정하면, 연결 과정에서 새 인덱스에 중복이 있는지 검사
- sort: 연결 시 열을 사전 순으로 정렬할지 여부를 결정/기본값은 False 
- copy: 기본값은 True/False로 설정하면, 가능한 경우 데이터 복사를 방지하여 성능을 개선



```python
pd.concat([test1,test2],1)
```

```python
pd.concat([test1,test2],join='outer')
```

```python
pd.concat([test1,test2],join='inner')
```
