# 데이터 프레임 전처리
- 지난주에 배웠던 내용
- 머신러닝 진행을 위한 코드 재실행

```python
# train_df와 test_df를 합쳐 combine 생성
train_df = train_df.drop(['Ticket','Cabin'], axis=1)
test_df = test_df.drop(['Ticket','Cabin'], axis=1)
combine = [train_df, test_df]
combine
```

## - Name 컬럼 전처리 

```python
for df in combine:
    df['Name_re'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=True)
```

```python
for df in combine:
    df['Name_re'] = df['Name_re'].replace(['Lady','Countess','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona', 'Capt'], 'Rare')
    df['Name_re'] = df['Name_re'].replace('Mlle','Miss')
    df['Name_re'] = df['Name_re'].replace('Ms','Miss')
    df['Name_re'] = df['Name_re'].replace('Mme','Miss')
```

```python
name_re = {'Mr':1, 'Miss':2, 'Mrs':3, 'Master':4,'Rare':5}

for df in combine:
    df['Name_re'] = df['Name_re'].map(name_re) 
    df['Name_re'] = df['Name_re'].fillna(0) # 결측치는 0으로
```

```python
#전처리 후 필요없는 칼럼 제거
train_df = train_df.drop(['Name','PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
combine = [train_df, test_df]
combine
```

```python
#sex 칼럼 전처리
for df in combine:
    df['Sex'] = df['Sex'].map({'female':1, 'male':0}).astype(int)
```

## - age 칼럼 전처리

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for df in combine:
    for i in range(0,2): # 성별을 나타냄
        for j in range(0,3): # Pclass를 나타냄
            # Sex, Pclass 기준으로 결측치를 제외한 Age 값을 guess_df에 저장
            guess_df = df[(df['Sex']==i) & (df['Pclass']==j+1)]['Age'].dropna()
            age_guess = guess_df.median()
            # age_guess을 반올림 하여 guess_ages에 저장 
            guess_ages[i,j] = (age_guess/0.5 + 0.5) * 0.5

    for i in range(0,2):
        for j in range(0,3):
            # Age의 결측치 채우기
            df.loc[(df.Age.isnull()) & (df.Sex ==i)&(df.Pclass==j+1), 'Age'] = guess_ages[i,j]
    # Age 컬럼을 int로 변환
    df['Age'] = df['Age'].astype(int)
```

```python
for df in combine:
    df.loc[df['Age'] <= 16,'Age'] = 0
    df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1 
    df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
    df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
    df.loc[df['Age'] > 64,'Age'] = 4
```

```python
# IsAlone 컬럼 전처리
for dataset in combine:
    # IsAlone 컬럼에 0으로 초기화 
    dataset['IsAlone'] = 0
    # 혼자 탑승한 사람은 IsAlone 컬럼 값을 1로 바꿔줌
    dataset.loc[dataset['FamilySize']==1 , 'IsAlone'] = 1
```

# 머신러닝 개념 정리
## 데이터 분석 과정
- 데이터 수집-> 데이터 전처리-> 데이터 모델링->데이터 모델링 평가->적용 및 전개
- 그동안 배웠던 내용이 데이터 전처리에 해당

## 앞으로 배울 내용
- 모델링을 통해 전저리하기 전과 후의 성능 비교(타이타닉 데이터로 진행)
- 독립변수,종속변수,피처들에 대한 이해가 필요함

## 연속형,범주형(이진,다중)데이터값에 따라 모델의 사용 방법이 달라짐
- 이진 분류모델링 - 로지스틱회귀, Decision Tree, RandomFroest, XGBoost, CatBoost, LightGBM 기타 등등..
- 연속형 회귀 모델링 - 선형회귀, 다중회귀, 다항회귀 등등, 분류 모델들도 다 회귀로도 모델링

## 데이터셋 나누기
- 데이터를 학습시키기위해 train,test,validation으로 데이터셋으 나눔
- train/test로 데이터를 나눈 이유: 데이터셋을 가지고 모델링을 하고 끝이 아니라 -> 이 데이터셋을 가지고 학습한 모델을 가지고 일반화를 통해서 다른 데이터셋이 들어왔을 때 동일하게 예측을 한다. **일반화** 일반화성능이 높아야 한다.

## 성능평가
- 이진 데이터: Accuracy, Recall, Precision, F1-score, AIC, BIC, AUC
- 연속형 데이터: MSE, RMSE 등 잔차에 대한 평가 지표
- 성능 평가 후 일반화 가능한지를 검증

## 과대적합,과소적합
- 과대적합 : 모델이 train 데이터에 너무 집착해서 학습하게 된 경우 -> train data 에서는 좋은 성능이 나오지만 -> test에서는 성능이 나쁘게 나옴
- 과소적합 : 모델이 너무 단순하거나, 데이터셋이 부족해서 -> train 성능이 잘 나오지 않고, test에서는 성능이 좋게 나옴
- train과 test dat간 성능의 간극을 줄이는 것이 가장 이상적인 모델임
 

```python
#x는 독립변수 y는 종속변수, x는 생존을 제외한 다른 피처, y는 생존, 정답
X_train = train_df.drop('Survived', axis=1)
Y_train = train_df['Survived']

X_test = test_df.drop('PassengerId',axis=1).copy()
```

```python
# 데이터 셋 개수 확인
X_train.shape, Y_train.shape, X_test.shape
```

```python
# 모델을 불러와서 진행
from sklearn.linear_model import LogisticRegression

# import 모델을 변수에 넣음
logreg= LogisticRegression() 
## 학습을 하기 위해 fit method를 이용해서 데이터를 학습시킴
logreg.fit(X_train, Y_train) # X_train, Y_train 모델에 학습 끝

# 예측값 -> test 데이터를 생존율 예측하는 정답을 받아야 함
# predict 사용해서 정답을 가지고 옴
# 미래값을 예측해야 하는 것이니 X_test test를 넣음
Y_pred_test=logreg.predict(X_test)
```
